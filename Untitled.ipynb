```python
import numpy as np
import nltk
from nltk.stem.porter import PorterStemmer 
from sklearn.feature_extraction.text import CountVectorizer 
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import ast
```

**Reading the datasets**

```python
movies = pd.read_csv('data/tmdb_5000_movies.csv')
movies.head()
movies.shape
```

```python
credits = pd.read_csv('data/tmdb_5000_credits.csv')
credits.head()
```

**merging the dataset**

```python
movies = movies.merge(credits,on='title')
movies.head()
```

**Feature selection**

```python
movies = movies [['id','title','genres','keywords','overview','cast','crew']]
movies.head()
```

**Data cleaning**


**Checking the null values**

```python
movies.isna().sum()
```

```python
movies = movies.dropna()  
```

**Checking the duplicate values**

```python
movies.duplicated().sum()
```

**Feature engineering**

```python
#function that display the format of data
def display_format(obj):
    return movies.iloc[0][obj]
print(f'Genres format: {display_format('genres')}\n')
print(f'keywords format: {display_format('keywords')}\n')
print(f'overview format: {display_format('overview')}\n')
```

**Conversion of data format**

```python
#this function is for converting the data format of genres and keywords column in the form of list
def convert_format(obj):
    return [i['name'] for i in ast.literal_eval(obj)]
movies['genres']= movies['genres'].apply(convert_format)
movies['keywords']= movies['keywords'].apply(convert_format)
```

```python
movies.info()
```

```python
#this function is for converting the data format of movies cast 
def convert_cast_format(obj):
    return [i['name'] for i in ast.literal_eval(obj)][:3]  #only extracting the top 3 cast names from each of the movie
movies['cast'] = movies['cast'].apply(convert_cast_format)    
```

```python
#this function is for converting the data format of movies crew
def convert_crew_format(obj):
    return [i['name'] for i in ast.literal_eval(obj) if i['job'] == 'Director']  #only extracting name of director from the movie, as we recommend the movie based on director too
movies['crew'] = movies['crew'].apply(convert_crew_format)    
```

**Changing the format of overview column**

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())  #converting the overview data which is in the form of sentence into list
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(' ','') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
movies['overview'] = movies['overview'].apply(lambda x:[i.replace(' ','') for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(' ','') for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(' ','') for i in x])
```

**Creation of movie tags**

```python
movies['tags'] = movies['genres'] + movies['keywords'] + movies['overview'] + movies['cast'] + movies['crew']
movies['tags']
```

**Converting the list into sentences**

```python
movies['tags'] = movies['tags'].apply(lambda x: ' '.join(x)) #also converting the words into  lowercase
movies['tags'] = movies['tags'].apply(lambda x: x.lower())
```

**Dataframe after feature engineering**

```python
movies = movies[['id','title','tags']]
movies
```

**Conversion of similar words into base word, which helps in count vectorization**

```python
ps = PorterStemmer()
def stemwords(text):
    ans = []
    for word in text.split(): #looping over the list of words, converted from sentence
        ans.append(ps.stem(word))
    return ' '.join(ans)  #converting the list back into sentence
    
movies['tags'] = movies['tags'].apply(stemwords)
```

**count based embeddings**

```python
cv =CountVectorizer(max_features = 5000, stop_words='english')  #here we are counting based on 5000 keywords from the movie tags, and based on this count we make the vector of tag for each movie
#and we skip the stop words such as a, the , and and so on.....
vectorized_tags = cv.fit_transform(movies['tags']).toarray()  #converting the sparse matrix into array
vectorized_tags
```

```python
similarity_data = cosine_similarity(vectorized_tags)
print("The top 5 similar movies' similarity value based on 1st indexed movie is:", sorted(list(enumerate(similarity_data[0])),reverse=True,key=lambda x:x[1])[1:5])  #sorting the list based on the similarity value not the index, cause using enumerate we get both the index as well as cosine similarity value
```

```python
movies[movies['title'] == 'Avatar'].index[0]
```

**Recommendation function**

```python
def recommend_movies(movie_name):
    movie_index = movies[movies['title'] == movie_name].index[0]  #finding the index of the movie in the similarity variable using the movie name
    movie_similarity_data = sorted(list(enumerate(similarity_data[movie_index])),reverse=True,key=lambda x:x[1])  #sorting the list based on the similarity value not the index in descending order, cause using enumerate we get both the index as well as cosine similarity value
    #finding the top 5 most similar movies
    for i in movie_similarity_data[1:6]:  #skipping the very first movie, as the first index is the same movie itself
        similar_movie_index = i[0]
        print(movies.iloc[similar_movie_index]['title'])  #then we print the similar movie title based on the movie index
recommend_movies('Avatar')        
```
